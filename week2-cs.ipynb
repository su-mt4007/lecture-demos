# Dataframes

```python
import pandas as pd
```

```python
# Read in data
# Via files or via URL
local_file_path = "./data/Winter_medals2022-11-03.csv"
url = "https://raw.githubusercontent.com/su-mt4007/data/refs/heads/main/Winter_medals2022-11-03.csv"
df = pd.read_csv(url)
df.head(20)
```

```python
# SELECT
cleaned_df = df[["Country", "Gold", "Silver", "Bronze"]]
cleaned_df
```

```python
# GROUPING & MUTATETING
# Get the total number by grouping
grouped_df = cleaned_df.groupby("Country").sum()
# Add new column for total.
grouped_df["Total"] = grouped_df.sum(axis=1)

grouped_df.head(5)
```

```python
# SORTING
## Sort countries by total number of medals
grouped_df.sort_values("Total", ascending=False).head()
```

```python
# FILTERING
## Get countries with more than 10 gold medals and 30 silver medals
grouped_df.loc[(grouped_df["Gold"]>10) & (grouped_df["Silver"]>=30)].sort_values("Total", ascending=False)
```

# Visualising Data & Plotting

```python
import matplotlib.pyplot as plt
```

## Simple line chart

```python
us_medals = df.loc[df["Country"] == "United States"]

x = us_medals["Year"]
y = us_medals["Gold"].cumsum()

plt.plot(x,y)
```

```python
country = "China"
df \
    .loc[df["Country"] == country] \
    [["Year", "Gold", "Silver", "Bronze"]] \
    .set_index("Year") \
    .plot.bar()
```

## Subplots

```python
fig, ax = plt.subplots(1,2)

ax[0].plot(x,y)
ax[0].set_title("This is a line chart")
ax[0].set_xlabel("x")
ax[0].set_ylabel("sin(x)")

ax[1].scatter(x,y)
ax[1].set_title("This is a scatter plot")
ax[1].set_xlabel("x")
```
